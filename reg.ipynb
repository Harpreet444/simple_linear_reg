```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
data_dic = {2600:550000,3000:565000,3200:610000,3600:680000,4000:725000}

# Convert dictionary to DataFrame
df = pd.DataFrame(list(data_dic.items()), columns=['Square_Feet', 'Price'])
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.Square_Feet,df.Price,color = 'g',marker = "*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Square_Feet']],df[['Price']])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
