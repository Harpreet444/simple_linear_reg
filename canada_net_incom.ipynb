```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(5)
```

```python
plt.scatter( df['year'],df['per capita income (US$)'], color='g', marker='*')
plt.ylabel('Per Capita Income (US$)')
plt.xlabel('Year')
plt.show()
```

```python
reg = linear_model.LinearRegression()
model = reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
plt.scatter(df['year'],df['per capita income (US$)'],color = 'r', marker='*')
plt.plot(df['year'],reg.predict(df[['year']]), color = 'g')
```
